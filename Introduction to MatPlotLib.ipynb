```python
import pandas as pd
heart_disease=pd.read_csv("heart-disease.csv")
heart_disease.head()
```

```python
heart_disease["age"].plot.hist(bins=10);
```

```python
heart_disease.plot.hist(figsize=(10,30),subplots=True);
```

```python
over_50=heart_disease[heart_disease["age"]>50]
over_50.head()
```

```python
over_50.plot(kind="scatter",
            x="age",
            y="chol",
            c="target");
```

```python
fix,ax=plt.subplots(figsize=(10,6))
over_50.plot(kind="scatter",
                   x="age",
                   y="chol",
                   c="target",
                   ax=ax);
ax.set_xlim([45,100])
```

```python
ax.set_xlim([45,100])
```

```python
fig,ax=plt.subplots(figsize=(10,6))
scatter=ax.scatter(x=over_50["age"],y=over_50["chol"],
                  c=over_50["target"]);
ax.set(title="Heart Disease and cholestral levels",
      xlabel="Age",
      ylabel="cholestoral");
ax.legend(*scatter.legend_elements(),title="target");
ax.axhline(over_50["chol"].mean(),linestyle="--");
```

```python
fig,(ax0,ax1)=plt.subplots(nrows=2,
                           ncols=1,
                           figsize=(10,10),
                          sharex=True)
scatter=ax0.scatter(x=over_50["age"],
                   y=over_50["chol"],
                   c=over_50["target"])
ax0.set(title="heart diseas",
       xlabel="age",
       ylabel="chol")
ax0.legend(*scatter.legend_elements(), title="target");
ax0.axhline(y=over_50["chol"].mean(),
           linestyle="--");
scatter=ax1.scatter(x=over_50["age"],
                   y=over_50["thalach"],
                   c=over_50["target"])
ax0.set(title="heart diseas",
       
       ylabel="thalach")
ax0.legend(*scatter.legend_elements(), title="target");
ax0.axhline(y=over_50["thalach"].mean(),
           linestyle="--");
fig.suptitle("Heart Disease",fontsize=16,fontweight="bold")
```
